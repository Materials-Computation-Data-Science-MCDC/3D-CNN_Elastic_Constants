```python
import os
import numpy as np
from skimage.color import rgb2gray
from skimage.transform import resize
from ase.calculators.vasp import VaspChargeDensity
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.colors import ListedColormap
import sys
import pylab
import time
```

## Generate color plots of all x-y planes in the z direction of the charge density

```python
def get_colormap(vcd,max_intensity,z_layer,style=None):
    """
    This function creates a 2D image of an x-y plane from the VASP
    charge density file.
    
    inputs:
        vcd: ase.calculators.vasp VaspChargeDensity input
        
        max_intensity (float):
            Sets the charge of anything above this value to be equal to this
            
        z_layer (int):
            The x-y plane of interest from your charge density file
        
        style (matplotlib listed colormap):
            User may define custom colormap for use if desired
    
    output:
        X (numpy array):
            3D numpy array of the chosen x-y charge density plane
    """
    
    # Define the style if no input is given
    if style is None:
        rainbow = mpl.colormaps['gist_ncar'].resampled(128)
        new_style = rainbow(np.linspace(0,1,128))
        new_style = new_style[48:,:]
        style = ListedColormap(new_style) 
    
    # Obtain only the density information from the vcd input
    density = vcd.chg[-1]

    # Set charges higher than the max intensity to the maximum
    # intensity everywhere in the plane of the image
    density[np.where(density > max_intensity)] = max_intensity

    # Define arbitrary meshgrid with shapes of charge density x-y inputs
    x = np.linspace(0,1,density.shape[0])
    y = np.linspace(0,1,density.shape[1])
    xv,yv = np.meshgrid(x,y)

    #Plotting the image
    fig,ax = plt.subplots(1,1,figsize = (3,3))
     
    ax.contourf(xv.T, yv.T, density[:,:,z_layer], 200, cmap = style)
    ax.axis('off')
    plt.axis('off')
    plt.tight_layout()
    plt.subplots_adjust(0,0,1,1,0,0)
    fig.add_axes(ax)
    fig.canvas.draw()

    # Save the figure as a numpy array and output as such
    X = np.array(fig.canvas.renderer._renderer)
    plt.close()
    return X
```

## Combine all layers of charge density into one 4D numpy array

```python
def create_3D_image_array(input_folder_path, output_folder_path):
    """
    This function combines all x-y charge density planes into one 3D 
    rectangular prism-like 4D numpy array.
    
    inputs:
        input_folder_path (str):
            Path to folders containing .npz files from get_colormap function
        
        output_folder_path (str):
            Path to where you want to save new combined 3D image arrays
            
    
    outputs:
        Saves the combined 4D numpy arrays to the output_folder_path
    """
    
    # Create the output subfolder if it doesn't exist
    if input_folder_path.split('/')[-1] != '':
        alloy = input_folder_path.split('/')[-1]
    else:
        alloy = input_folder_path.split('/')[-2]
    # print(input_folder_path.split('/'))
    output_subfolder_path = os.path.join(output_folder_path, alloy)
    os.makedirs(output_subfolder_path, exist_ok=True)
    
    # List to store the arrays
    arrays_list = []

    # Iterate over all files in the input folder
    for subfolder_name in os.listdir(input_folder_path):
        subfolder_path = os.path.join(input_folder_path, subfolder_name)

        # Check if it's a directory
        if os.path.isdir(subfolder_path):
            # List of files in the subfolder
            subfolder_files = os.listdir(subfolder_path)

            # Check if the subfolder contains .npz files
            if any(file.endswith('.npz') for file in subfolder_files):
                # List to store arrays for this subfolder
                subfolder_arrays_list = []

                # Iterate over .npz files in the subfolder
                for filename in subfolder_files:
                    if filename.endswith('.npz'):
                        # Construct the full file path
                        file_path = os.path.join(subfolder_path, filename)

                        # Load the .npz file
                        npz_file = np.load(file_path)

                        # Extract the array using the key
                        array = npz_file['a']

                        # Append the array to the subfolder list
                        subfolder_arrays_list.append(array)

                # Convert the list of arrays to a numpy array for this subfolder
                subfolder_final_array = np.array(subfolder_arrays_list)

                # Save the numpy array to the output folder with the subfolder name
                # output_subfolder_path = os.path.join(output_folder_path, subfolder_name)
                # os.makedirs(output_subfolder_path, exist_ok=True)
                np.save(os.path.join(output_subfolder_path, f'{subfolder_name}_array.npy'), subfolder_final_array)

    print("Arrays saved successfully.")
```

## Resize and recolor images to grayscale if desired

```python
def resize_image(source_folder_path, target_folder_path, resize_shape=None):
    """
    This function resizes the 4D numpy arrays, and makes them into grayscale
    allowing for them to be reduced to 3D numpy arrays for less computational
    resources.
    
    inputs:
        source_folder_path (str):
            Folder where 4D numpy array files are stored (generated from create_3D_image_array)
            
        target_folder_path (str):
            Folder where new resized 3D gray-scale images are to be stored
            
        resize_shape (tuple):
            x-y shape to resize images to
            
    outputs:
        3D gray-scale numpy array files stored to target_folder_path
    """
    
    # Create the target folder if it doesn't exist
    os.makedirs(target_folder_path, exist_ok=True)

    # Resize factor for images (300 to 100)
    if resize_shape is None:
        resize_shape = (100, 100)

    # Iterate over all files in the source folder
    for filename in os.listdir(source_folder_path):
        if filename.endswith('_array.npy'):
            # Construct the full file paths
            source_file_path = os.path.join(source_folder_path, filename)
            target_file_path = os.path.join(target_folder_path, filename)

            # Load the numpy array
            original_array = np.load(source_file_path)

            # Extract the first three channels
            three_channels_array = original_array[..., :3]

            # Convert to grayscale
            grayscale_array = rgb2gray(three_channels_array)

            # Resize the array
            resized_array = resize(grayscale_array, resize_shape, anti_aliasing=True)

            # Expand dimensions to (X, 100, 100, 1)
            resized_array = np.expand_dims(resized_array, axis=-1)

            # Save the processed array to the target folder
            np.save(target_file_path, resized_array)

    print(f"Processing complete. Arrays saved in the target folder: {target_folder_path}.")
```

```python
# Loop through and generate the numpy arrays for each z layer within a structure's charge density
keys = ['1-NiCu']
for key in keys:
    # Directory containing VASP charge density files
    directory = os.fsencode('/project/daidhy/mcdc/nlinton/rd-hea_data/elastic-modulus/dft/1-binary/%s/1-chgcar_files/'%key)

    os.chdir('/scratch/nlinton/elastic_constants/image_prep/')
    for file in os.listdir(directory):
        f_name = os.fsdecode(file)
        if not f_name.endswith('.py'):
            chg_density_file=os.fsdecode(directory)+f_name

            # Check if VASP charge density file has information stored
            if os.stat(chg_density_file).st_size != 0:
         
                # Read in the charge density file 
                vcd = VaspChargeDensity(chg_density_file)
                charge_density = vcd.chg
                
                # Get shape of charge density in each x, y, z directions
                a = charge_density[0]
                x,y,z  = a.shape
    
                # Create unique folder name for your structure (unique to your naming convention)
                fol_name = f_name.split('-CHGCAR')[0]
    
                # Try creating a folder, if it exists move onto the next
                try:
                    os.makedirs('%s_color1/%s'%(key,fol_name)) #, exist_ok=True)
                    # os.system('sleep 0.1')
                    os.chdir('%s_color1/'%key+fol_name)
                    
                    # Loop through all z ranges
                    for i in range(0,z):
                        new = str(i+1).zfill(4) # create a string for file naming convention of 0001 - 9999
                        
                        # Run the get_colormap function to get final output
                        X = get_colormap(vcd, 0.4, i,)
                        
                        # Save the numpy array as a .npz file
                        np.savez_compressed('%s'%new,a=X)
                    os.chdir('../../')
                except:
                    print('%s is already done' %f_name)

    # os.system('tar -cf %s_color1.tar %s_color1/'%(key,key))
```
